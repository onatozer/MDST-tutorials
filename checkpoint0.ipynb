```python
# fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 30 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number

for i in range(1,30):
    if(i % 3 != 0 and i % 5 != 0):
        print(i)
        continue
    if(i % 3 == 0):
        print("fizz", end = "")
    if(i % 5 == 0):
        print("buzz", end = "")
    print()
    
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd

df = pd.read_csv("../data/starbucks.csv")
type(df)
print(df)
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 
tableOne = df.iloc[::40, [3,11,12]]
print(tableOne)
```

```python
# select all rows with more than and including 400 calories
tableTwo = df[df["calories"] >= 400]
print(tableTwo)
```

```python
# select all rows whose vitamin c content is higher than the iron content
tableThree = df[df["vitamin c"] >= df["iron"]]
print(tableThree)
```

```python
# create a new column containing the caffeine per calories of each drink
df["caffeine per calories"] = df["caffeine"] / df["calories"] 

print(df)
```

```python
# what is the average calorie across all items?

total_calories = 0
dCalories = df["calories"]
for i in dCalories:
    total_calories += i

print("Average calories is: ", total_calories/242)
```

```python
# how many different categories of beverages are there?
total_categories = 1
previous_category = "coffee"
dCategories = df["beverage_category"]
for i in dCategories:
    if(previous_category != i):
        total_categories +=1
        previous_category = i

print("There are:", total_categories, "categories")
```

```python
# what is the average # calories for each beverage category?
print("The average number of calories per beverage category is", (total_calories/total_categories))
```

```python
# plot the distribution of the number of calories in drinks with a histogram
df["calories"].plot.hist(edgecolor='black', alpha=0.8, title="Calorie Distrbution across drinks", xlabel = "calories")
```

```python
# plot calories against total fat with a scatterplot

df.plot.scatter(x="calories", y="total fat", title="Total fat vs calorie content")
```
